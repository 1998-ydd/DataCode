```python
df['Xdist'] = ((df['X1'] - df['X2'])**2 + (df['Y1'] - df['Y2'])**2)**0.5
df['Ydist'] = ((df['X1'] - df['X3'])**2 + (df['Y1'] - df['X3'])**2)**0.5
df['Single'] = ((df['X1'])**2 + (df['Y1'])**2)**0.5
df['Single-mean'] = ((df['X1']-df['X1'].mean())**2 + (df['Y1']-df['Y1'].mean())**2)**0.5
df['Single2'] = ((df['X2'])**2 + (df['Y2'])**2)**0.5
df['Single2-mean'] = ((df['X2']-df['X2'].mean())**2 + (df['Y2']-df['Y2'].mean())**2)**0.5
df['Single3'] = ((df['X3'])**2 + (df['X3'])**2)**0.5
df['Single3-mean'] = ((df['X3']-df['X3'].mean())**2 + (df['X3']-df['X3'].mean())**2)**0.5
```
